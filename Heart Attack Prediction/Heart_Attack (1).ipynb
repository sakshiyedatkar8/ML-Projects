```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#to read dataset
dataset = pd.read_csv('C:/Users/SAKSHI/Downloads/heart_dataset.csv')
#head() returns first 5 rows
dataset.head()
```

```python
#bar plot of target attribute
plt.bar(dataset['target'].unique(),dataset['target'].value_counts(),color=['red','yellow'])
```

```python
#create dummies
dataset = pd.get_dummies(dataset, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])

#Normalization/Scaling
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])

#View dataset after scaling
dataset.head()
```

```python

X = dataset.drop('target',axis=1)
y = dataset['target']

#training testing split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

#import KNN
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=8)
classifier.fit(X_train,y_train)
classifier.predict(X_test)

#Accuracy
print(classifier.score(X_test,y_test)*100)
```
